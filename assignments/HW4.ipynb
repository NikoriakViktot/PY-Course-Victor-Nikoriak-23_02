Маніпуляції з рядками

Напишіть програму на Python, яка отримує рядок, що складається з перших 2 та останніх 2 символів заданого рядка. Якщо довжина рядка менша за 2, поверніть замість порожнього рядка.

Зразок рядка:'helloworld'

Очікуваний результат:'held'

Зразок рядка:'my'

Очікуваний результат:'mymy'

Зразок рядка:'x'

Очікуваний результат:Empty String

Поради:

Використання вбудованої функції len()для вхідного рядка
Використовуйте позитивну індексацію для отримання перших символів рядка та негативну індексацію для отримання останніх символів.
 

```python
while True:
    x = input("name: ")
    if x.lower() == "exit":
        print("Програму завершено.")
        break
        

    if len(x) > 2:
        print(x[:2] + x[-2:])
    elif len(x) == 2:
        print(x * 2)
    else:  # тобто len(x) < 2
        print("Empty")
```

Створіть програму, яка перевіряє, чи рядок має правильний формат для номера телефону. Програма повинна перевірити, чи рядок містить лише числові символи та чи має довжину лише 10 символів. Виведіть відповідне повідомлення залежно від результату обчислення рядка.

```python
user_N=input ("Введіть номер телефону +380xxxxxxxxx: ")
if user_N.isdigit() and len(user_N)==9:
    print("Номер телефону правильний")
else:
    print("Номер телефону неправильний")
```

Програма математичної вікторини.

Напишіть програму, яка запитує відповідь на математичний вираз, перевіряє, чи правий користувач, чи ні, а потім відповідає відповідним повідомленням.

```python
task=input("Напишіть розв'язок прикладу: 5+5")
correct_answer=10
answer=input ("Ваша відповідь: ")
if answer.isdigit() and int(answer)==correct_answer:
    print("Відповідь правильна")
else:
    print("Відповідь неправильна")
```

Напишіть програму, яка містить змінну з вашим іменем (у нижньому регістрі), а потім запитує ваше ім'я як вхідні дані. Програма повинна перевіряти, чи введене ім'я дорівнює збереженому імені, навіть якщо задане ім'я має інший регістр, наприклад, якщо ви ввели «Антон», а збережене ім'я «антон», вона повинна повернути True.

```python
save_name="світлана"
user_name=input("Введіть ваше ім'я: ")
if user_name.lower() == save_name.lower():
    print("Привіт, " + user_name + "!")
else:
    print("Ви не є " + save_name + ".")
```
