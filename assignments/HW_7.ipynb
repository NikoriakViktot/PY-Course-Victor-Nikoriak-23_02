Створіть програму, яка на вхід має деяке речення (рядок) і повертає словник, що містить усі унікальні слова як ключі та кількість входжень як значення. 

```python
input= str("Світло переможе темряву")
words= input.split()
print(words)
dict= {}
for i in words:
    if i in dict:
        dict[i] += 1
    else:
        dict[i] = 1
print(dict)
```

Обчисліть загальну ціну запасу, де загальна ціна дорівнює сумі ціни товару, помноженої на кількість цього конкретного товару.

Код має повертати словник із сумами цін за типами товарів.




```python
stock = {
    "banana": 6,
    "apple": 0,
    "orange": 32,
    "pear": 15
}
prices = {
    "banana": 4,
    "apple": 2,
    "orange": 1.5,
    "pear": 3
}
total_price = {item: stock[item] * prices[item] for item in stock if item in prices}
print(total_price)
```

Вправа на розуміння списку

Використайте списочне сприйняття, щоб створити список, що містить кортежі (i, j), де 'i' змінюється від 1 до 10, а 'j' відповідає 'i' у квадраті.

 

```python
i= list(range(1, 11))
j= [x**2 for x in i]

my_list= list(zip(i, j))

print(i)
print(j)
print(my_list)
```

Створити лист із днями тижня.
В один рядок (ну або як завжди) створіть словник виду: {1: "Понеділок", 2:...
Також в один рядок або як вдасться створити зворотний словник {"Понеділок": 1,

```python
from calendar import day_name

week= list(day_name)
week_dict = {i: day for i, day in enumerate(week, 1)}
reverse_dict = {day: i for i, day in enumerate(week, 1)}

print(week_dict)
print(reverse_dict)
```
