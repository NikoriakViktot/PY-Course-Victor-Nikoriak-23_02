Напишіть програму на Python, яка отримує найбільше число зі списку випадкових чисел довжиною 10.

Обмеження: використовувати лише цикл while та модуль random для генерації чисел

```python
import random

number_list=[]
i=0
while i<10:
    number_list.append(random.randint(0,100))
    i+=1
print(number_list)
max_number=number_list[0]

k=0
while k<len(number_list):
    if number_list[k]>max_number:
        max_number=number_list[k]
    k+=1

print(max_number)
```

Створіть 2 списки довжиною 10 з випадковими цілими числами від 1 до 10 та третій список, що містить спільні цілі числа між двома початковими списками без будь-яких дублікатів.

Обмеження: використовувати лише цикл while та модуль random для генерації чисел

```python
import random

number_list1=[]
number_list2=[]
i=0
while i<10:
    number_list1.append(random.randint(0,10))
    number_list2.append(random.randint(0,10))
    i+=1
print(number_list1)
print(number_list2)

list_common = []
j = 0
while j < len(number_list1):
    if number_list1[j] in number_list2 and number_list1[j] not in list_common:
        list_common.append(number_list1[j])
    j += 1

print(list_common)
```
