```python
Завдання 1

Напишіть декоратор, який виводить функцію з переданими їй аргументами.

ПРИМІТКА! Слід вивести функцію, а не результат її виконання!

 Наприклад:
 "додати викликані з 4, 5"
 
def logger(функція):
    пропуск
 @logger

def додати(x, y):
    повернути x + y
 @logger
def square_all(*args):
    повернути [arg ** 2 для аргументу в аргументах]
```

```python
def logger(func):
    def wrapper(*args, **kwargs):
        # перетворюємо аргументи у рядок
        args_str = ", ".join(map(str, args))
        kwargs_str = ", ".join(f"{k}={v}" for k, v in kwargs.items())
        all_args = ", ".join(filter(None, [args_str, kwargs_str]))

        # друкуємо повідомлення про виклик
        print(f"{func.__name__} викликано з {all_args}")
        return func(*args, **kwargs)  # викликаємо саму функцію
    return wrapper


@logger
def додати(x, y):
    return x + y

@logger
def square_all(*args):
    return [arg ** 2 for arg in args]


# Виклики
додати(4, 5)
square_all(2, 3, 4)
```

Напишіть декоратор, який приймає список стоп-слів та замінює їх на * всередині декорованої функції.
 '''

'''

```python
def stop_words(words: list):
    def decorator(func):
        def wrapper(*args, **kwargs):
            text = func(*args, **kwargs)
            for w in words:
                text = text.replace(w, "*")
            return text
        return wrapper
    return decorator
@stop_words(['пепсі', 'BMW'])
def create_slogan(name: str) -> str:
    return f"{name} п'є пепсі у своєму новенькому BMW!"
# Перевірка
assert create_slogan("Стів") == "Стів п'є * у своєму новенькому *!"
print("Усі тести пройдено успішно ✅")
```

Завдання 3

Напишіть декоратор 'arg_rules', який перевіряє аргументи, передані функції.
Декоратор повинен приймати 3 аргументи:
максимальна_довжина: 15
тип_: рядок
містить: [] - список символів, які має містити аргумент
Якщо деякі з перевірок правил повертають False, функція повинна повернути False та вивести причину невдачі; інакше повернути результат.
'''
def arg_rules(type_: тип, max_length: int, contains: список):
    пропуск
@arg_rules(type_=str, max_length=15, contains=['05', '@'])
def create_slogan(назва: рядок) -> рядок:
    поверни f"{name} п'є пепсі у своєму новенькому BMW!"
стверджувати, що create_slogan('johndoe05@gmail.com') є хибним
assert create_slogan('S@SH05') == 'S@SH05 п'є пепсі у своєму новенькому BMW!

```python
def arg_rules(type_: type, max_length: int, contains: list):
    def decorator(func):
        def wrapper(arg):
            # перевірка типу
            if not isinstance(arg, type_):
                print(f"❌ Помилка: аргумент має бути типу {type_.__name__}, але отримано {type(arg).__name__}")
                return False

            # перевірка довжини
            if len(arg) > max_length:
                print(f"❌ Помилка: довжина аргументу перевищує {max_length} символів")
                return False

            # перевірка наявності символів
            for item in contains:
                if item not in arg:
                    print(f"❌ Помилка: аргумент повинен містити '{item}'")
                    return False

            # якщо всі перевірки пройдено
            return func(arg)
        return wrapper
    return decorator


@arg_rules(type_=str, max_length=15, contains=['05', '@'])
def create_slogan(name: str) -> str:
    return f"{name} п'є пепсі у своєму новенькому BMW!"


# Перевірки
assert create_slogan('johndoe05@gmail.com') is False
assert create_slogan('S@SH05') == "S@SH05 п'є пепсі у своєму новенькому BMW!"

print("Усі тести пройдено успішно ✅")
```
