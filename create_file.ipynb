Напишіть скрипт, який створює новий вихідний файл з назвою myfile.txt та записує в нього рядок "Hello file world!". Потім напишіть ще один скрипт, який відкриває myfile.txt, зчитує та друкує його вміст. Запустіть два скрипти з командного рядка системи. 

Чи відображається новий файл у каталозі, де ви запускали свої скрипти? 

Що робити, якщо додати інший шлях до імені файлу, переданого для відкриття?

Примітка: методи запису у файл не додають символи нового рядка до ваших рядків; додайте явний символ "\n" в кінці рядка, якщо ви хочете повністю завершити рядок у файлі.

 

```python
# create_file.py
with open("myfile.txt", "w") as f:
    f.write("Hello file world!\n")
    f.write("This is a new file created by Python.\n")
```

```python
# read_file.py
with open("myfile.txt", "r") as f:
    content = f.read()
    print(content)
```
